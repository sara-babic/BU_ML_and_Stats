# 3. Using Hierarchical Bayesian Models in Your Own Research

AGN time-series models will generally be described with paramaters for a damped random walk (DRW) and maybe a quasi-periodic operator (QPO):

- **Damped Random Walk (DRW):**
  - can model the intrinsic random/stochastic, time-correlated noise aspect of AGN lightcurves
  - observational trace: red-noise
  - parameters (2): characteristic timescale $\tau$ and signal variance amplitude $\sigma_\rm{DRW}^2$
- **Quasi-Periodic Oscillator (QPO) :**
  - can model the repeatting patterns of binary SMBH AGN lightcurves 
  - observational traces: quasi-periodic peak
  - parameters (3): natural frequency of oscillation $\omega$, quality factor $Q$ which relates the damping rate and the sharpness of peak in frequency spectrum (e.g. high $Q$ means less damping (slower decay of exponential term) and sharper feature in curve), the signal variance $\sigma_\rm{QPO}^2$.
